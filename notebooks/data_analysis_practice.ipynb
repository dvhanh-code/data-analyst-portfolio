```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```
df = pd.read_csv(
    "/Users/anhdang/PycharmProjects/data-analyst-portfolio/datasets/Superstore.csv",
    encoding="ISO-8859-1"
)

df.head()
```

```
df.info()
```


### Revenue by Category

```
sales_by_category = df.groupby("Category")["Sales"].sum()

sales_by_category.plot(kind="bar")

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.savefig("../images/sales_by_category.png")
plt.show()
```

### Sales Over Time

```
df["Order Date"] = pd.to_datetime(df["Order Date"])

sales_over_time = df.groupby("Order Date")["Sales"].sum()

sales_over_time.plot()

plt.title("Sales Over Time")
plt.savefig("../images/sales_over_time.png")
plt.show()
```

### Sales by Region

```
sales_by_region = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(8,5))

sales_by_region.plot(kind="bar", color="orange")

plt.title("Total Sales by Region")
plt.ylabel("Sales")

plt.savefig("../images/sales_by_region.png", dpi=300, bbox_inches="tight")

plt.show()
```

### Profit by Category

```
profit_by_category = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))

profit_by_category.plot(kind="bar", color="green")

plt.title("Profit by Category")
plt.ylabel("Profit")

plt.savefig("../images/profit_by_category.png", dpi=300, bbox_inches="tight")

plt.show()
```

### Discount vs Profit

```
plt.figure(figsize=(8,5))

sns.scatterplot(x="Discount", y="Profit", data=df)

plt.title("Discount vs Profit")

plt.savefig("../images/discount_vs_profit.png", dpi=300, bbox_inches="tight")

plt.show()
```
